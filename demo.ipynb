```python
# With the help of yolov7 detection of the dog and cropping the dog object from the image

with torch.no_grad():
     dog1 = detect(img_path="sample_image/golden-retriever-dog-21668976.jpg")
     dog2 = detect(img_path="sample_image/golden-retriever_Shutterstock_Olena-Brodetska-760x505.jpg")
     dog3 = detect(img_path="sample_image/bigstock-black-dalmatian-dog-outdoors-i-96150509.png")
```

```python
from extractor.inference import inference
```

```python
# Feature extracted from the each cropped dog image
# And saved in the form of dog vector of size 512 vector

dog1_feature = inference(img=dog1[0])
dog2_feature = inference(img=dog2[0])
dog3_feature = inference(img=dog3[0])
```

```python
from similarity_score import similarity
```

```python
# Similarity score between 2 same breed dogs

similarity(dog1_feature[0],dog2_feature[0])
```

```python
# Similarity score between 2 different breed dogs

similarity(dog1_feature[0],dog3_feature[0])
```
